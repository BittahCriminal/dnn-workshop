# Image Classification with MLServer

This sample trains a multiclass linear model using the 
`microsoftml:rx_logistic_regression` algorithm to distinguish between fish, helicopter and fighter jet images.
Feature vactors for the images are extracted using a pre-trained model supplied with the `microsoftml` library.
The multiclass training task uses the feature vectors of the images from the training set to learn how to classify these images.

The procedure for training the model has the following steps:

1. Locate the images to use in the training set and get their feature vectors.
2. Label the images in the training set by type.
3. Featurize the images with ResNet
4. Train a multiclass classifier using the `microsoftml:rx_logistic_regression` algorithm.
5. Specify a new image not in the training set to classify and use the trained model 
  to predict its type.

```python
import os

try:
    root = os.path.dirname(__file__)
except NameError:
    # __file__ does not exist in a notebook
    root = "."

# An absolute path must be used if the current folder
# is not the script's one.
image_location = os.path.abspath(os.path.join(root, "Data", "Pictures"))
```

Read all images in the local data folder.
The last subfolder name corresponds to the class the image belongs to. 

```python
import glob
images = glob.glob(image_location + "/**/*.jpg", recursive=True)
images_type = [img.replace("\\", "/").split("/")[-2] for img in images]
print(list(zip(images_type, images))[0])
print(list(zip(images_type, images))[-1])
```

We need a numeric label for the images. 
we put images and their corresponding label in a dataframe.

```python
import pandas
images_df = pandas.DataFrame(data=dict(image=images, image_type=images_type))

label_int = {'Fish': 0., 'Helicopter': 1., 'FighterJet':2., 'Flower': 3.}
images_df["Label"] = images_df["image_type"].apply(lambda t: label_int[t])
images_df = images_df[["Label", "image_type", "image"]]
print(images_df)
```

Display the first and last images

```python
import matplotlib.pyplot as plt
from PIL import Image
fig, ax = plt.subplots(1, 2)
ax[0].imshow(Image.open(images_df.loc[0, "image"]))
ax[1].imshow(Image.open(images_df.loc[images_df.shape[0]-1, "image"]))
plt.show()
```

Pick a random image for testing, the rest for training

```python
import random
h = random.randint(0, images_df.shape[0]-1)
train_df = images_df.drop(h).reset_index(drop=True)
test_df = images_df[h:h+1].reset_index(drop=True)
print(test_df)

fig, ax = plt.subplots(1, 1)
ax.imshow(Image.open(test_df.loc[0, "image"]))
plt.show()
```

## The Key Point
We train a multi-class classifier using multinomial logistic regression (aka softmax LR).
The classifier uses feature vectors as an input, it does not understand images.
In order to create the feature vectors, we use a convolution neural network from the pre-trained model library. In this case, we opted for Resnet50.
Featurization is treated as a transform to be executed before the classification.
The transform is actually composed of 4 steps:
1. Load the iamge from the training set
2. Resize the image to the format supported by the trained network
3. Extract the actual pixel values
4. Featurize with the DNN to generate vectors

```python
from microsoftml import rx_featurize, load_image, resize_image, extract_pixels, featurize_image
from microsoftml import rx_logistic_regression

image_model = rx_logistic_regression(formula="Label~Features", data=train_df, 
                                     method="multiClass", ml_transforms=[
                                         load_image(cols=dict(Features="image")),
                                         resize_image(cols="Features", width=224, height=224),
                                         extract_pixels(cols="Features"),
                                         featurize_image(cols="Features", dnn_model="Resnet50")])
```

Classify the test set

```python
from microsoftml import rx_predict
prediction = rx_predict(image_model, data=test_df)
print(prediction)

###############################
# The highest score gives the predicted label.
import numpy as np
label_str = {v: k for k, v in label_int.items()}
predicted_classes = np.argmax(prediction.as_matrix(), axis=1)
print(predicted_classes)
print(label_str[predicted_classes[0]])
```

```python
import pickle
filename = './model.pkl'
pickle.dump(image_model, open(filename, mode='wb'))
```

## Now try to publish to MLServer


```python
# -- Import the DeployClient and MLServer classes from the azureml-model-management-sdk package.
from azureml.deploy import DeployClient
from azureml.deploy.server import MLServer

# -- Define the location of the ML Server --
# -- for local onebox for Machine Learning Server: http://localhost:12800
# -- Replace with connection details to your instance of ML Server. 
HOST = 'http://<your cluster name here>.<your region name here>.cloudapp.azure.com'
context = ('admin', '<your password here>')
client = DeployClient(HOST, use=MLServer, auth=context)
```

```python
# --Define an `init` function to handle service initialization --
def init():
    import pandas as pd
    import numpy as np
    from microsoftml import rx_predict
    from microsoftml import rx_featurize, load_image, resize_image, extract_pixels, featurize_image
    from microsoftml import rx_logistic_regression
    import urllib.request as url
    import os
    import string
    import random
```

```python
def classify_image(imagefile):
    import pandas as pd
    import numpy as np
    from microsoftml import rx_predict
    from microsoftml import rx_featurize, load_image, resize_image, extract_pixels, featurize_image
    from microsoftml import rx_logistic_regression
    import urllib.request as url
    import os
    import string
    import random
    
    #generate a random file name, then copy image
    #there is a bug in the 9.3 release where a url will not error out, but will not work either
    downloaded="./downloaded.jpg"+''.join(random.choice(string.ascii_uppercase + string.digits) for _ in range(8))
    url.urlretrieve (imagefile, downloaded)
    
    newdata = pd.DataFrame({'Label':[-1], 'image_type':['unknown'], 'image': [downloaded]})

    prediction = rx_predict(model=image_model, data=newdata)
    
    # -- save some files to demonstrate the ability to return file artifacts --
    answer = np.argmax(prediction.as_matrix(), axis=1)
    
    answer.tofile('answer.csv')
    
    #delete tmp file
    os.remove(downloaded)
    
    # return prediction
    return answer
```

```python
# the service name + version must be unique. Either increment the version or choose a unique name
service_name = 'imgService'

service = client.service(service_name)\
        .version('1.2')\
        .code_fn(classify_image, init)\
        .inputs(imagefile=str)\
        .outputs(answer=pandas.DataFrame)\
        .models(image_model=image_model)\
        .description('My first image classification model')\
        .artifacts(['answer.csv'])\
        .deploy()
```

```python
print(help(service))
```

```python
# extract capabilities
service.capabilities()
```

```python
# call the service. Since we are in the same session, we need note authenticate again
result = service.classify_image('https://gmexperimentstorage.blob.core.windows.net/data/Data/Pictures/Flower/Flower3.jpg')

# Return the named output `answer` as defined during publishing and print the result
print(result.output('answer'))
print(label_str[int(result.output('answer')[0][0])])
```
